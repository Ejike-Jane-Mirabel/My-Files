```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('sample_pivot.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Units'].fillna(np.mean(df['Units']), inplace=True)
```

```python
df
```

```python
#pivot is used for statistical analysis between categorical column using numerical columns as the factor, 
#it gives better insight and understanding of your table.
 
df_pv = pd.pivot_table(df, index='Region', columns='Type', values=['Units','Sales'])
df_pv
```

```python
df_pv = pd.pivot_table(df, index='Region', columns='Type', values=['Units','Sales'],aggfunc=max)
df_pv
```

```python
df_pv = pd.pivot_table(df, index='Region', columns='Type', values=['Units','Sales'],aggfunc='count')
df_pv
```

```python
df['Date'] = pd.to_datetime(df['Date']) # change the date in the column to the actual date
df.info()
```

```python
df[(df['Date'] >= pd.Timestamp(2020,4,15)) & (df['Date'] <= pd.Timestamp(2020,10,30))]
```

```python
east = df[df['Region']== 'East']
west = df[df['Region']== 'West']
north = df[df['Region']== 'North']
south = df[df['Region']== 'South']
```

```python
east
```

```python
south
```

```python
east.merge(west,how='right')
```

```python
east.merge(west,how='left')
```

```python
east.merge(west,how='inner')
```

```python
east.merge(west,how='outer')
```

```python
alldf = pd.concat([east,west,north,south])
alldf
```

```python
alldf.to_csv('clothing.csv', index=False)
```

```python
alldf.to_excel('clothing.xlsx', index=False, sheet_name='Sales')
```

```python
with pd.ExcelWriter('ALLSALES.xlsx', engine='xlsxwriter') as writer:
    east.to_excel(writer, sheet_name='EAST',index=False, engine='xlsxwriter')
    west.to_excel(writer, sheet_name='WEST',index=False, engine='xlsxwriter')
    north.to_excel(writer, sheet_name='NORTH',index=False, engine='xlsxwriter')
    south.to_excel(writer, sheet_name='SOUTH',index=False, engine='xlsxwriter')
```
